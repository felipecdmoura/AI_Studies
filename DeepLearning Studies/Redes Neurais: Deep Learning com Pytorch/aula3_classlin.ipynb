```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(199)

X, Y = make_classification(n_features=2, n_redundant=0, n_informative=1,
                             n_clusters_per_class=1)

plt.scatter(X[:, 0], X[:, 1], marker='o', c=Y,
           edgecolor='k')
```

```python




def plotmodel(w1,w2,b):

    plt.scatter(X[:, 0], X[:, 1], marker='o', c=Y,
           edgecolor='k')


    xmin, xmax = plt.gca().get_xlim()
    ymin, ymax = plt.gca().get_ylim()

    x = np.linspace(-5, 4, 50)
    y = (-w1*x -b)/w2

    plt.axvline(0, -10, 10, color='k', linewidth=1)
    plt.axhline(0, -10, 10, color='k', linewidth=1)

    plt.plot(x,y)
    plt.xlim(xmin, xmax)
    plt.ylim(ymin, ymax)

w1 = -1
w2 = 6
b = 0
plotmodel(w1,w2,b)
p = X[1]
plt.plot(p[0], p[1], marker="^", color='r', markersize=30)
```

```python
def classify(p, w1,w2,b):
    line = w1*p[0] + w2*p[1] + b
    if line >= 0:
        return 1, 'yellow'
    else:
        return 0, 'blue'

count = 0

for k in range(len(X)):
    rot, _ = classify(X[k], w1, w2, b)

    if rot == Y[k]:
        count += 1
print(X[1])
print(w1*p[0] + w2*p[1] + b)
print(count/len(X)*100);


```

```python
import torch 
from torch import nn


perc = nn.Linear(2,1)
perc.weight = nn.Parameter(torch.Tensor([[-1, 6]]))
perc.bias = nn.Parameter(torch.Tensor([0]))



inp = torch.from_numpy(X[1])
out = perc(inp.float())

print(out)
```
